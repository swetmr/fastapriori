# Generalized K-Itemset Demo (k=2,3,4,5)

Demonstrates `find_itemsets()` — a single parametric function that handles any itemset size from pairs through quintuplets, with Apriori pruning at each level.

```python
import time
import tracemalloc

import numpy as np
import pandas as pd
from faker import Faker

from fastapriori import find_associations, find_itemsets, find_triplets
```

## 1. Generate Dummy Dataset

```python
NUM_TRANSACTIONS = 50_000
NUM_ITEMS = 200
MEAN_ITEMS_PER_TXN = 6

Faker.seed(42)
np.random.seed(42)
fake = Faker()

items = [fake.bothify(text="??####").upper() for _ in range(NUM_ITEMS)]

# Zipf-skewed popularity: a few items are very common
weights = 1.0 / np.arange(1, NUM_ITEMS + 1) ** 0.8
weights /= weights.sum()

rows = []
for txn_id in range(1, NUM_TRANSACTIONS + 1):
    n_items = np.random.poisson(lam=MEAN_ITEMS_PER_TXN - 1) + 1
    n_items = min(n_items, NUM_ITEMS)
    chosen = np.random.choice(items, size=n_items, replace=False, p=weights)
    for item in chosen:
        rows.append((txn_id, item))

df = pd.DataFrame(rows, columns=["Transaction ID", "Item"])

print(f"Rows:              {len(df):,}")
print(f"Transactions:      {df['Transaction ID'].nunique():,}")
print(f"Unique items:      {df['Item'].nunique():,}")
print(f"Mean items/txn:    {df.groupby('Transaction ID')['Item'].count().mean():.1f}")
```

## 2. Pairwise Associations (k=2)

Start with `find_associations()` to get the frequent pairs. These are the foundation for Apriori pruning at higher k.

```python
t0 = time.perf_counter()
pairs = find_associations(
    df,
    transaction_col="Transaction ID",
    item_col="Item",
    min_support=0.001,
    min_confidence=0,
    show_progress=True,
)
t_pairs = time.perf_counter() - t0

print(f"\nTime:  {t_pairs:.2f}s")
print(f"Rules: {len(pairs):,}")
pairs.sort_values("lift", ascending=False).head(10)
```

## 3. Triplets (k=3)

Using `find_itemsets(k=3)` with pair-level Apriori pruning.

```python
t0 = time.perf_counter()
triplets = find_itemsets(
    df,
    transaction_col="Transaction ID",
    item_col="Item",
    k=3,
    min_support=0.001,
    min_confidence=0.1,
    frequent_lower=pairs,
    show_progress=True,
)
t_triplets = time.perf_counter() - t0

print(f"\nTime:  {t_triplets:.2f}s")
print(f"Rules: {len(triplets):,}")
triplets.sort_values("lift", ascending=False).head(10)
```

### Verify: `find_triplets()` wrapper produces identical output

```python
triplets_old = find_triplets(
    df,
    transaction_col="Transaction ID",
    item_col="Item",
    min_support=0.001,
    min_confidence=0.1,
    frequent_pairs=pairs,
)
print(f"find_triplets == find_itemsets(k=3): {triplets_old.equals(triplets)}")
```

## 4. Quadruplets (k=4)

Pass the triplet results as `frequent_lower` for pruning and confidence computation.

```python
t0 = time.perf_counter()
quads = find_itemsets(
    df,
    transaction_col="Transaction ID",
    item_col="Item",
    k=4,
    min_support=0.001,
    min_confidence=0.1,
    frequent_lower=triplets,
    show_progress=True,
)
t_quads = time.perf_counter() - t0

print(f"\nTime:  {t_quads:.2f}s")
print(f"Rules: {len(quads):,}")
quads.sort_values("lift", ascending=False).head(10)
```

## 5. Quintuplets (k=5)

Pass the quadruplet results as `frequent_lower`.

```python
t0 = time.perf_counter()
quints = find_itemsets(
    df,
    transaction_col="Transaction ID",
    item_col="Item",
    k=5,
    min_support=0.001,
    min_confidence=0.1,
    frequent_lower=quads,
    show_progress=True,
)
t_quints = time.perf_counter() - t0

print(f"\nTime:  {t_quints:.2f}s")
print(f"Rules: {len(quints):,}")
quints.sort_values("lift", ascending=False).head(10)
```

## 6. Summary

```python
summary = pd.DataFrame([
    {"k": 2, "Time (s)": round(t_pairs, 2), "Rules": len(pairs)},
    {"k": 3, "Time (s)": round(t_triplets, 2), "Rules": len(triplets)},
    {"k": 4, "Time (s)": round(t_quads, 2), "Rules": len(quads)},
    {"k": 5, "Time (s)": round(t_quints, 2), "Rules": len(quints)},
])
summary
```

## 7. Sanity Checks

```python
for name, result, k in [("triplets", triplets, 3), ("quads", quads, 4), ("quints", quints, 5)]:
    ant_cols = [f"antecedent_{i}" for i in range(1, k)]
    checks = {
        "has correct columns": set(ant_cols + ["consequent", "instances", "support", "confidence", "lift"]).issubset(result.columns),
        "support <= 1.0": (result["support"] <= 1.0).all(),
        "confidence <= 1.0 + eps": (result["confidence"] <= 1.0 + 1e-6).all(),
        "lift > 0": (result["lift"] > 0).all(),
        "min_support respected": (result["support"] >= 0.001).all(),
        "min_confidence respected": (result["confidence"] >= 0.1).all(),
    }
    status = "PASS" if all(checks.values()) else "FAIL"
    print(f"k={k} ({name}): {status}")
    for check, ok in checks.items():
        if not ok:
            print(f"  FAILED: {check}")
```

## 8. Example: Chained Pipeline

The typical workflow chains each level's output into the next for pruning:

```python
# Full pipeline: k=2 -> k=3 -> k=4 -> k=5
k2 = find_associations(df, "Transaction ID", "Item", min_support=0.001, min_confidence=0)
k3 = find_itemsets(df, "Transaction ID", "Item", k=3, frequent_lower=k2, min_confidence=0.1)
k4 = find_itemsets(df, "Transaction ID", "Item", k=4, frequent_lower=k3, min_confidence=0.1)
k5 = find_itemsets(df, "Transaction ID", "Item", k=5, frequent_lower=k4, min_confidence=0.1)

print(f"k=2: {len(k2):,} rules")
print(f"k=3: {len(k3):,} rules")
print(f"k=4: {len(k4):,} rules")
print(f"k=5: {len(k5):,} rules")
```
