# fastapriori — Test Notebook

Test with `faker.csv` (uniform distribution) and a skewed dataset that produces meaningful associations.

```python
import pandas as pd
import numpy as np
from fastapriori import find_associations
```

## Test 1: faker.csv (uniform distribution)

With 5K uniformly distributed items across 200K transactions, max confidence is ~0.025 and support rounds to 0.0. This is expected — uniform distribution has no meaningful co-occurrence patterns. We run with no filters to confirm the algorithm works.

```python
df_faker = pd.read_csv("../faker.csv")
print(f"Rows: {len(df_faker):,} | Transactions: {df_faker['Transaction ID'].nunique():,} | Items: {df_faker['Item Number'].nunique():,}")

result_uniform = find_associations(df_faker, "Transaction ID", "Item Number", min_support=None, min_confidence=None, show_progress=True)
print(f"\nResult rows (no filter): {len(result_uniform):,}")
print(f"Support  — min: {result_uniform['support'].min()}, max: {result_uniform['support'].max()}")
print(f"Confidence — min: {result_uniform['confidence'].min():.4f}, max: {result_uniform['confidence'].max():.4f}")
print(f"Lift — min: {result_uniform['lift'].min():.4f}, max: {result_uniform['lift'].max():.4f}")
result_uniform.head(10)
```

## Test 2: Skewed distribution (realistic co-occurrence)

To get meaningful support/confidence values, we need items with skewed popularity — some items appear in many transactions and frequently co-occur. This uses a Zipf-like distribution for item selection.

```python
np.random.seed(42)

NUM_ITEMS = 500
NUM_TRANSACTIONS = 50_000
item_pool = [f"ITEM-{i:04d}" for i in range(NUM_ITEMS)]

# Zipf-like weights: a few items are very popular, most are rare
weights = 1.0 / np.arange(1, NUM_ITEMS + 1) ** 0.8
weights /= weights.sum()

rows = []
for txn_id in range(NUM_TRANSACTIONS):
    n_items = np.random.poisson(lam=5) + 1
    n_items = min(n_items, NUM_ITEMS)
    items = np.random.choice(item_pool, size=n_items, replace=False, p=weights)
    for item in items:
        rows.append((txn_id, item))

df_skewed = pd.DataFrame(rows, columns=["Transaction ID", "Item Number"])
print(f"Rows: {len(df_skewed):,} | Transactions: {df_skewed['Transaction ID'].nunique():,} | Items: {df_skewed['Item Number'].nunique():,}")
print(f"Mean items/txn: {len(df_skewed) / df_skewed['Transaction ID'].nunique():.1f}")

# Show item popularity distribution
item_counts = df_skewed["Item Number"].value_counts()
print(f"\nItem popularity — top 5: {item_counts.head().tolist()}, bottom 5: {item_counts.tail().tolist()}")
```

```python
df_skewed.to_csv("faker_skewed.csv", index= False)
```

```python
result_skewed = find_associations(
    df_skewed, "Transaction ID", "Item Number",
    min_support=0.01, min_confidence=0.1, show_progress=True,
)
print(f"Result rows: {len(result_skewed):,}")
print(f"Support  — min: {result_skewed['support'].min()}, max: {result_skewed['support'].max()}")
print(f"Confidence — min: {result_skewed['confidence'].min()}, max: {result_skewed['confidence'].max()}")
print(f"Lift — min: {result_skewed['lift'].min():.4f}, max: {result_skewed['lift'].max():.4f}")
result_skewed.head(20)
```

## Test 3: Sanity checks

```python
# No self-associations
assert (result_skewed["item_A"] == result_skewed["item_B"]).sum() == 0, "FAIL: self-associations found"
print("✓ No self-associations")

# No confidence > 1
assert (result_skewed["confidence"] > 1.0).sum() == 0, "FAIL: confidence > 1"
print("✓ All confidence <= 1.0")

# Support and confidence thresholds respected
assert result_skewed["support"].min() >= 0.01, "FAIL: support below threshold"
print("✓ All support >= 0.01")

assert result_skewed["confidence"].min() >= 0.1, "FAIL: confidence below threshold"
print("✓ All confidence >= 0.1")

# Lift > 0
assert (result_skewed["lift"] > 0).all(), "FAIL: non-positive lift"
print("✓ All lift > 0")

print("\nAll sanity checks passed!")
```
