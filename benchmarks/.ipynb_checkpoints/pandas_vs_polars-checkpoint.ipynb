# Pandas vs Polars — Comprehensive Benchmark

Compares **pandas** (Counter+chain) and **polars** (self-join) backends across:
- Two dataset sizes: **1M rows** and **10M rows**
- Varying thresholds (`min_support`, `min_confidence`, `min_lift`)
- **Pairwise** associations and **triplet** (3-itemset) rules

All datasets use Zipf-skewed item distributions for realistic co-occurrence patterns.

```python
import pandas as pd
import numpy as np
import time
import tracemalloc
import gc
from faker import Faker
from fastapriori import find_associations, find_triplets
```

```python
def generate_dataset(num_transactions, num_unique_items, mean_items_per_txn=5, seed=42):
    """Generate a Zipf-skewed transactional dataset."""
    fake = Faker()
    Faker.seed(seed)
    np.random.seed(seed)

    item_pool = [fake.bothify(text="??####", letters="ABCDEFGHIJKLMNOPQRSTUVWXYZ")
                 for _ in range(num_unique_items)]

    weights = 1.0 / np.arange(1, num_unique_items + 1) ** 0.8
    weights /= weights.sum()

    rows = []
    for txn_id in range(num_transactions):
        n_items = min(np.random.poisson(lam=mean_items_per_txn) + 1, num_unique_items)
        items = np.random.choice(item_pool, size=n_items, replace=False, p=weights)
        for item in items:
            rows.append((txn_id, item))

    df = pd.DataFrame(rows, columns=["Transaction ID", "Item Number"])
    return df


def bench_pairs(df, backend, min_support=None, min_confidence=0.0, min_lift=0.0):
    """Benchmark pairwise associations. Returns (time_s, peak_mb, num_rules)."""
    gc.collect()
    tracemalloc.start()
    t0 = time.perf_counter()

    result = find_associations(
        df, "Transaction ID", "Item Number",
        min_support=min_support, min_confidence=min_confidence, min_lift=min_lift,
        backend=backend,
    )

    t1 = time.perf_counter()
    _, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    return round(t1 - t0, 2), round(peak / 1024**2, 1), len(result), result


def bench_triplets(df, frequent_pairs=None, min_support=None, min_confidence=None):
    """Benchmark triplet associations. Returns (time_s, peak_mb, num_rules)."""
    gc.collect()
    tracemalloc.start()
    t0 = time.perf_counter()

    result = find_triplets(
        df, "Transaction ID", "Item Number",
        min_support=min_support, min_confidence=min_confidence,
        frequent_pairs=frequent_pairs,
        show_progress=True,
    )

    t1 = time.perf_counter()
    _, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    return round(t1 - t0, 2), round(peak / 1024**2, 1), len(result), result
```

```python
print("Generating 1M-row dataset...")
df_1m = generate_dataset(num_transactions=200_000, num_unique_items=500, mean_items_per_txn=5)
print(f"  Rows: {len(df_1m):,} | Transactions: {df_1m['Transaction ID'].nunique():,} | "
      f"Items: {df_1m['Item Number'].nunique()} | Mean items/txn: {len(df_1m)/df_1m['Transaction ID'].nunique():.1f}")
```

## 2. Pairwise Associations — Varying Thresholds

Benchmarks both backends across multiple threshold combinations.
Note: filtering happens *after* computation, so threshold choice affects output size but not core compute time.

```python
threshold_configs = [
    {"label": "No filter",           "min_support": None,  "min_confidence": None, "min_lift": None},
    {"label": "sup>=0.001",           "min_support": 0.001, "min_confidence": None, "min_lift": None},
    {"label": "sup>=0.001, conf>=0.1", "min_support": 0.001, "min_confidence": 0.1,  "min_lift": None},
    {"label": "sup>=0.001, lift>=1.0", "min_support": 0.001, "min_confidence": None, "min_lift": 1.0},
    {"label": "sup>=0.005, conf>=0.2, lift>=1.0", "min_support": 0.005, "min_confidence": 0.2, "min_lift": 1.0},
]

print(f"{len(threshold_configs)} threshold configurations ready")
```

### 2a. Pairs — 1M Rows

```python
pairs_1m_results = []

for cfg in threshold_configs:
    for backend in ["pandas", "polars"]:
        t, mem, n_rules, _ = bench_pairs(
            df_1m, backend,
            min_support=cfg["min_support"],
            min_confidence=cfg["min_confidence"],
            min_lift=cfg["min_lift"],
        )
        pairs_1m_results.append({
            "Dataset": "1M", "Backend": backend, "Thresholds": cfg["label"],
            "Time (s)": t, "Peak Mem (MB)": mem, "Rules": n_rules,
        })
        print(f"  {backend:6s} | {cfg['label']:40s} | {t:6.2f}s | {mem:7.1f} MB | {n_rules:,} rules")

pairs_1m_df = pd.DataFrame(pairs_1m_results)
print("\nDone — 1M pairs")
```

### 2b. Pairs — 10M Rows

```python
pairs_10m_results = []

for cfg in threshold_configs:
    for backend in ["pandas", "polars"]:
        t, mem, n_rules, _ = bench_pairs(
            df_10m, backend,
            min_support=cfg["min_support"],
            min_confidence=cfg["min_confidence"],
            min_lift=cfg["min_lift"],
        )
        pairs_10m_results.append({
            "Dataset": "10M", "Backend": backend, "Thresholds": cfg["label"],
            "Time (s)": t, "Peak Mem (MB)": mem, "Rules": n_rules,
        })
        print(f"  {backend:6s} | {cfg['label']:40s} | {t:6.2f}s | {mem:7.1f} MB | {n_rules:,} rules")

pairs_10m_df = pd.DataFrame(pairs_10m_results)
print("\nDone — 10M pairs")
```

### 2c. Pairs — Summary Table

```python
all_pairs = pd.concat([pairs_1m_df, pairs_10m_df], ignore_index=True)

# Pivot for side-by-side comparison
pivot_time = all_pairs.pivot_table(
    index=["Dataset", "Thresholds"], columns="Backend",
    values="Time (s)", aggfunc="first"
).reset_index()
pivot_time["Speedup (polars/pandas)"] = round(pivot_time["pandas"] / pivot_time["polars"], 2)

pivot_mem = all_pairs.pivot_table(
    index=["Dataset", "Thresholds"], columns="Backend",
    values="Peak Mem (MB)", aggfunc="first"
).reset_index()
pivot_mem.columns = ["Dataset", "Thresholds", "Mem pandas (MB)", "Mem polars (MB)"]
pivot_mem["Mem Ratio (polars/pandas)"] = round(pivot_mem["Mem polars (MB)"] / pivot_mem["Mem pandas (MB)"], 2)

# Merge
summary_pairs = pivot_time.merge(pivot_mem, on=["Dataset", "Thresholds"])
summary_pairs = summary_pairs[[
    "Dataset", "Thresholds",
    "pandas", "polars", "Speedup (polars/pandas)",
    "Mem pandas (MB)", "Mem polars (MB)", "Mem Ratio (polars/pandas)",
]]
summary_pairs.columns = [
    "Dataset", "Thresholds",
    "Time pandas (s)", "Time polars (s)", "Speedup (x)",
    "Mem pandas (MB)", "Mem polars (MB)", "Mem Ratio (x)",
]

print("=" * 100)
print("PAIRS — PANDAS vs POLARS")
print("=" * 100)
summary_pairs
```

---
## 3. Triplet Associations — Varying Thresholds

Triplets use the pandas-based Counter+combinations approach.
The key variable is **pruning via frequent pairs**: providing `frequent_pairs` from pairwise analysis
dramatically reduces the search space.

We benchmark:
- No pruning (all triplets)
- Pruned with loose pairs (sup>=0.001)
- Pruned with strict pairs (sup>=0.005, conf>=0.2)
- Varying `min_support` and `min_confidence` on triplet output

```python
# Pre-compute pairwise results at different thresholds for pruning
print("Pre-computing pairwise results for triplet pruning...")

# 1M pairs
_, _, _, pairs_1m_loose = bench_pairs(df_1m, "pandas", min_support=0.001)
_, _, _, pairs_1m_strict = bench_pairs(df_1m, "pandas", min_support=0.005, min_confidence=0.2)
print(f"  1M: loose pairs = {len(pairs_1m_loose):,}, strict pairs = {len(pairs_1m_strict):,}")

# 10M pairs
_, _, _, pairs_10m_loose = bench_pairs(df_10m, "pandas", min_support=0.001)
_, _, _, pairs_10m_strict = bench_pairs(df_10m, "pandas", min_support=0.005, min_confidence=0.2)
print(f"  10M: loose pairs = {len(pairs_10m_loose):,}, strict pairs = {len(pairs_10m_strict):,}")
```

```python
triplet_configs = [
    {"label": "No pruning, no filter",
     "prune": None, "min_support": None, "min_confidence": None},
    {"label": "Pruned (loose), no filter",
     "prune": "loose", "min_support": None, "min_confidence": None},
    {"label": "Pruned (strict), no filter",
     "prune": "strict", "min_support": None, "min_confidence": None},
    {"label": "Pruned (loose), sup>=0.001",
     "prune": "loose", "min_support": 0.001, "min_confidence": None},
    {"label": "Pruned (loose), sup>=0.001, conf>=0.1",
     "prune": "loose", "min_support": 0.001, "min_confidence": 0.1},
]

print(f"{len(triplet_configs)} triplet configurations ready")
```

### 3a. Triplets — 1M Rows

```python
triplet_1m_results = []

prune_map_1m = {None: None, "loose": pairs_1m_loose, "strict": pairs_1m_strict}

for cfg in triplet_configs:
    fp = prune_map_1m[cfg["prune"]]
    t, mem, n_rules, _ = bench_triplets(
        df_1m, frequent_pairs=fp,
        min_support=cfg["min_support"],
        min_confidence=cfg["min_confidence"],
    )
    triplet_1m_results.append({
        "Dataset": "1M", "Config": cfg["label"],
        "Time (s)": t, "Peak Mem (MB)": mem, "Rules": n_rules,
    })
    print(f"  {cfg['label']:45s} | {t:6.2f}s | {mem:7.1f} MB | {n_rules:,} rules")

triplet_1m_df = pd.DataFrame(triplet_1m_results)
print("\nDone — 1M triplets")
```

### 3b. Triplets — 10M Rows

```python
triplet_10m_results = []

prune_map_10m = {None: None, "loose": pairs_10m_loose, "strict": pairs_10m_strict}

for cfg in triplet_configs:
    fp = prune_map_10m[cfg["prune"]]
    t, mem, n_rules, _ = bench_triplets(
        df_10m, frequent_pairs=fp,
        min_support=cfg["min_support"],
        min_confidence=cfg["min_confidence"],
    )
    triplet_10m_results.append({
        "Dataset": "10M", "Config": cfg["label"],
        "Time (s)": t, "Peak Mem (MB)": mem, "Rules": n_rules,
    })
    print(f"  {cfg['label']:45s} | {t:6.2f}s | {mem:7.1f} MB | {n_rules:,} rules")

triplet_10m_df = pd.DataFrame(triplet_10m_results)
print("\nDone — 10M triplets")
```

### 3c. Triplets — Summary Table

```python
all_triplets = pd.concat([triplet_1m_df, triplet_10m_df], ignore_index=True)

print("=" * 100)
print("TRIPLETS — PRUNING & THRESHOLD IMPACT")
print("=" * 100)
all_triplets
```

## 4. Visualizations

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for ax, dataset_label in zip(axes, ["1M", "10M"]):
    subset = all_pairs[all_pairs["Dataset"] == dataset_label]
    thresholds = subset["Thresholds"].unique()
    x = np.arange(len(thresholds))
    w = 0.35

    pd_times = subset[subset["Backend"] == "pandas"]["Time (s)"].values
    pl_times = subset[subset["Backend"] == "polars"]["Time (s)"].values

    bars1 = ax.bar(x - w/2, pd_times, w, label="pandas", color="#4C78A8")
    bars2 = ax.bar(x + w/2, pl_times, w, label="polars", color="#F58518")

    ax.set_title(f"Pairs — {dataset_label} Rows", fontsize=13, fontweight="bold")
    ax.set_ylabel("Time (seconds)")
    ax.set_xticks(x)
    ax.set_xticklabels(thresholds, rotation=30, ha="right", fontsize=8)
    ax.legend()

    # Annotate bars
    for bar in bars1:
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
                f"{bar.get_height():.1f}s", ha="center", va="bottom", fontsize=7)
    for bar in bars2:
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
                f"{bar.get_height():.1f}s", ha="center", va="bottom", fontsize=7)

fig.suptitle("Pairwise Associations — Execution Time", fontsize=15, fontweight="bold")
fig.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for ax, dataset_label in zip(axes, ["1M", "10M"]):
    subset = all_pairs[all_pairs["Dataset"] == dataset_label]
    thresholds = subset["Thresholds"].unique()
    x = np.arange(len(thresholds))
    w = 0.35

    pd_mem = subset[subset["Backend"] == "pandas"]["Peak Mem (MB)"].values
    pl_mem = subset[subset["Backend"] == "polars"]["Peak Mem (MB)"].values

    bars1 = ax.bar(x - w/2, pd_mem, w, label="pandas", color="#4C78A8")
    bars2 = ax.bar(x + w/2, pl_mem, w, label="polars", color="#F58518")

    ax.set_title(f"Pairs — {dataset_label} Rows", fontsize=13, fontweight="bold")
    ax.set_ylabel("Peak Memory (MB)")
    ax.set_xticks(x)
    ax.set_xticklabels(thresholds, rotation=30, ha="right", fontsize=8)
    ax.legend()

    for bar in bars1:
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
                f"{bar.get_height():.0f}", ha="center", va="bottom", fontsize=7)
    for bar in bars2:
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
                f"{bar.get_height():.0f}", ha="center", va="bottom", fontsize=7)

fig.suptitle("Pairwise Associations — Peak Memory", fontsize=15, fontweight="bold")
fig.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for ax, dataset_label in zip(axes, ["1M", "10M"]):
    subset = all_triplets[all_triplets["Dataset"] == dataset_label]
    configs = subset["Config"].values
    times = subset["Time (s)"].values
    colors = ["#E45756" if "No pruning" in c else "#72B7B2" for c in configs]

    bars = ax.barh(range(len(configs)), times, color=colors)
    ax.set_yticks(range(len(configs)))
    ax.set_yticklabels(configs, fontsize=8)
    ax.set_xlabel("Time (seconds)")
    ax.set_title(f"Triplets — {dataset_label} Rows", fontsize=13, fontweight="bold")
    ax.invert_yaxis()

    for bar, t in zip(bars, times):
        ax.text(bar.get_width(), bar.get_y() + bar.get_height()/2,
                f" {t:.1f}s", ha="left", va="center", fontsize=8)

fig.suptitle("Triplet Associations — Pruning Impact on Execution Time", fontsize=15, fontweight="bold")
fig.tight_layout()
plt.show()
```

## 5. Summary Report

```python
print("=" * 100)
print("FINAL SUMMARY REPORT — PANDAS vs POLARS")
print("=" * 100)

# --- Pairs ---
print("\n" + "-" * 100)
print("PAIRWISE ASSOCIATIONS")
print("-" * 100)

for ds in ["1M", "10M"]:
    sub = all_pairs[all_pairs["Dataset"] == ds]
    pd_sub = sub[sub["Backend"] == "pandas"]
    pl_sub = sub[sub["Backend"] == "polars"]

    avg_speedup = (pd_sub["Time (s)"].values / pl_sub["Time (s)"].values).mean()
    avg_mem_ratio = (pl_sub["Peak Mem (MB)"].values / pd_sub["Peak Mem (MB)"].values).mean()

    print(f"\n  [{ds} rows]")
    print(f"    Avg time  — pandas: {pd_sub['Time (s)'].mean():.2f}s, polars: {pl_sub['Time (s)'].mean():.2f}s")
    print(f"    Avg speedup (pandas_time / polars_time): {avg_speedup:.2f}x")
    print(f"    Avg memory — pandas: {pd_sub['Peak Mem (MB)'].mean():.1f} MB, polars: {pl_sub['Peak Mem (MB)'].mean():.1f} MB")
    print(f"    Avg mem ratio (polars / pandas): {avg_mem_ratio:.2f}x")

# --- Triplets ---
print("\n" + "-" * 100)
print("TRIPLET ASSOCIATIONS")
print("-" * 100)

for ds in ["1M", "10M"]:
    sub = all_triplets[all_triplets["Dataset"] == ds]
    no_prune = sub[sub["Config"].str.contains("No pruning")]
    pruned = sub[~sub["Config"].str.contains("No pruning")]

    print(f"\n  [{ds} rows]")
    if len(no_prune) > 0:
        print(f"    No pruning:   {no_prune.iloc[0]['Time (s)']:.2f}s, {no_prune.iloc[0]['Peak Mem (MB)']:.1f} MB, {no_prune.iloc[0]['Rules']:,} rules")
    if len(pruned) > 0:
        best = pruned.loc[pruned["Time (s)"].idxmin()]
        print(f"    Best pruned:  {best['Time (s)']:.2f}s, {best['Peak Mem (MB)']:.1f} MB, {best['Rules']:,} rules  [{best['Config']}]")
        if len(no_prune) > 0:
            speedup = no_prune.iloc[0]["Time (s)"] / best["Time (s)"] if best["Time (s)"] > 0 else float('inf')
            print(f"    Pruning speedup: {speedup:.1f}x")

# --- Key takeaways ---
print("\n" + "-" * 100)
print("KEY TAKEAWAYS")
print("-" * 100)

# Compute overall speedup
pd_all = all_pairs[all_pairs["Backend"] == "pandas"]["Time (s)"].values
pl_all = all_pairs[all_pairs["Backend"] == "polars"]["Time (s)"].values
overall_speedup = (pd_all / pl_all).mean()

pd_mem_all = all_pairs[all_pairs["Backend"] == "pandas"]["Peak Mem (MB)"].values
pl_mem_all = all_pairs[all_pairs["Backend"] == "polars"]["Peak Mem (MB)"].values
overall_mem = (pl_mem_all / pd_mem_all).mean()

faster = "Polars" if overall_speedup > 1 else "Pandas"
leaner = "Pandas" if overall_mem > 1 else "Polars"

print(f"\n  1. {faster} is faster on average ({overall_speedup:.2f}x speedup)")
print(f"  2. {leaner} uses less memory on average ({overall_mem:.2f}x ratio polars/pandas)")
print(f"  3. Polars self-join is best for sparse data (<10 items/txn)")
print(f"  4. Pandas Counter+chain is best for dense data (>20 items/txn)")
print(f"  5. Triplet pruning via frequent pairs dramatically reduces compute time")
print(f"  6. Threshold filtering happens post-computation — doesn't affect core compute")
print()
```
