# fastapriori Benchmark: fastapriori vs efficient-apriori

Compares execution time and peak memory usage on a 1M-row synthetic dataset.

```python
import pandas as pd
import numpy as np
import time
import tracemalloc
from faker import Faker
import fastapriori
from efficient_apriori import apriori
```

## 1. Generate Dummy Dataset (1M rows)

```python
fake = Faker()
Faker.seed(42)
np.random.seed(42)

NUM_UNIQUE_ITEMS = 500
NUM_TRANSACTIONS = 200_000

# Generate pool of realistic item numbers
item_pool = [fake.bothify(text="??####", letters="ABCDEFGHIJKLMNOPQRSTUVWXYZ") for _ in range(NUM_UNIQUE_ITEMS)]

# Zipf-like weights: a few items are very popular, most are rare
# This creates realistic co-occurrence patterns with meaningful support/confidence
weights = 1.0 / np.arange(1, NUM_UNIQUE_ITEMS + 1) ** 0.8
weights /= weights.sum()

# Build transactions with Poisson-distributed item counts (mean ~5)
rows = []
for txn_id in range(NUM_TRANSACTIONS):
    n_items = min(np.random.poisson(lam=5) + 1, NUM_UNIQUE_ITEMS)
    items = np.random.choice(item_pool, size=n_items, replace=False, p=weights)
    for item in items:
        rows.append((txn_id, item))

df = pd.DataFrame(rows, columns=["Transaction ID", "Item Number"])
df.to_csv("faker.csv", index=False)
```

```python
df= pd.read_csv("faker.csv")
```

```python
print(f"Generated {len(df):,} rows")
print(f"Unique transactions: {df['Transaction ID'].nunique():,}")
print(f"Unique items: {df['Item Number'].nunique():,}")
print(f"Mean items per transaction: {len(df) / df['Transaction ID'].nunique():.1f}")

# Show item popularity skew
item_counts = df["Item Number"].value_counts()
print(f"\nItem popularity — top 5: {item_counts.head().tolist()}, bottom 5: {item_counts.tail().tolist()}")
```

```python
df.head(10)
```

## 2. Benchmark fastapriori

```python
tracemalloc.start()

t0 = time.perf_counter()
result_fast = fastapriori.find_associations(
    df,
    transaction_col="Transaction ID",
    item_col="Item Number",
    min_support=0.0,
    min_confidence=0.01,
    show_progress=True,
)
t1 = time.perf_counter()

_, peak_fast = tracemalloc.get_traced_memory()
tracemalloc.stop()

time_fast = t1 - t0
print(f"fastapriori: {time_fast:.2f}s | peak memory: {peak_fast / 1024**2:.1f} MB")
print(f"Result rows: {len(result_fast):,}")
```

```python
result_fast.head(5)
```

## 3. Benchmark efficient-apriori

```python
# Convert to list-of-tuples format required by efficient-apriori
transactions_list = df.groupby("Transaction ID")["Item Number"].apply(tuple).tolist()
print(f"Prepared {len(transactions_list):,} transactions for efficient-apriori")
```

```python
tracemalloc.start()

t0 = time.perf_counter()
try:
    itemsets, rules = apriori(
        transactions_list,
        min_support=0.0,
        min_confidence=0.01,
        max_length=2,
    )
    t1 = time.perf_counter()
    _, peak_ea = tracemalloc.get_traced_memory()
    tracemalloc.stop()

    time_ea = t1 - t0
    num_rules = len(rules)
    print(f"efficient-apriori: {time_ea:.2f}s | peak memory: {peak_ea / 1024**2:.1f} MB")
    print(f"Rules found: {num_rules}")
except Exception as e:
    t1 = time.perf_counter()
    _, peak_ea = tracemalloc.get_traced_memory()
    tracemalloc.stop()

    time_ea = t1 - t0
    num_rules = 0
    print(f"efficient-apriori failed after {time_ea:.2f}s: {e}")
    print(f"Peak memory before failure: {peak_ea / 1024**2:.1f} MB")
```

## 4. Results Comparison

```python
comparison = pd.DataFrame({
    "Method": ["fastapriori", "efficient-apriori"],
    "Time (s)": [round(time_fast, 2), round(time_ea, 2)],
    "Peak Memory (MB)": [round(peak_fast / 1024**2, 1), round(peak_ea / 1024**2, 1)],
    "Output Rows/Rules": [len(result_fast), num_rules],
})

# Speedup relative to the slower method
max_time = comparison["Time (s)"].max()
comparison["Speedup"] = round(max_time / comparison["Time (s)"], 1)

comparison
```

## 5. Analysis

**Why fastapriori is faster for large datasets:**

- **No candidate generation:** Traditional Apriori generates candidate itemsets and prunes based on support. fastapriori skips this entirely — it directly counts co-occurrences using `Counter` + `itertools.chain` over pre-built set lookups.
- **Vectorized pandas operations:** Groupby, explode, and arithmetic are handled by pandas/numpy internals, minimizing Python-level loops.
- **Set-based lookups:** Transaction-to-items mapping uses Python dicts of sets, giving O(1) lookups per transaction.

**Where efficient-apriori has different strengths:**

- Supports arbitrary itemset sizes (triplets, quadruplets, etc.) out of the box
- Pruning via the Apriori property can be advantageous when `min_support` is high (filters early, examines fewer candidates)
- Well-suited for smaller datasets where candidate generation overhead is negligible
