```python
import numpy as np
import pandas as pd
from collections import Counter, defaultdict
import time
from itertools import combinations, chain
```

```python
start= time.perf_counter()
```

```python
df= pd.read_csv('tblBundlingInputMar24.csv', on_bad_lines='skip', sep= "|")
```

```python
df.head()
```

```python
df.shape
```

```python
df.dropna(inplace= True)
```

```python
df.isna().sum()
```

```python
#df.head(30)
```

```python
df.shape
```

```python
df['UID'].nunique()
```

```python
df['Current PN'].nunique()
```

```python
%%time

inv= df.groupby('UID')['Current PN'].apply(set).to_frame().reset_index()
```

```python
inv
```

```python
inv['len']= inv['Current PN'].apply(len)
```

```python
inv_dict= dict(zip(inv['UID'], inv['Current PN']))
```

```python
len(inv_dict)
```

```python
part= df.groupby('Current PN')['UID'].apply(set).to_frame().reset_index()
```

```python
part['len']= part['UID'].apply(len)
```

```python
part['Current PN']= part['Current PN'].astype(str)
```

```python
%%time

part['Counter']= part['UID'].apply(lambda x: Counter(chain(*(inv_dict[i] for i in x))).items())
```

```python
part
```

```python
part['Len Final']= part['Counter'].apply(len)
```

```python
part_final= part[part['Len Final']>0]
```

```python
part_final.shape
```

```python
#part_final
```

## Explode columns

```python
bun= part_final[['Current PN', 'len', 'Counter']].explode(['Counter']).reset_index(drop=True)
```

```python
bun['Part']= bun['Counter'].apply(lambda x: x[0])
```

```python
bun['Part Occurence']= bun['Counter'].apply(lambda x: x[1])
```

```python
bun['Association %age']= bun['Part Occurence']/bun['len']
```

```python
bun['Association %age']= np.round(bun['Association %age'], 2)
```

```python
bun_sub= bun[(bun['Current PN']!= bun['Part']) & (bun['Association %age']>=0.10)]
```

```python
bun_sub.drop(columns= ['Counter', 'len'], inplace= True)
```

```python
#bun_sub.reset_index(inplace= True)
```

```python
bun_sub.rename(columns= {'Current PN': 'Master Part Number', 'Part Occurence': 'No. of Instances'}, inplace= True)
```

```python
bun_sub
```

```python
bun_sub.to_csv('Bundling Full Revised.csv', index= False)
```

```python
end= time.perf_counter()
```

```python
print('Time taken is', (end-start)/60, 'minutes')
```
