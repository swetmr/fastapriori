```python
def get_all_parts(x):
    
    parts= defaultdict(lambda:[])
    
    for k,v in zip(x['Part Number Att'].values.tolist(), x['Month Description'].values) :
        
        parts[k].append(v)
        
    parts= {k:set(v) for k,v in parts.items()}
    
    return parts
```

```python
## Logic for taking 1 date: average, latest
```

```python
def get_bundle(x):
    
    com= list(combinations(x.keys(), 2))
    
    valid= []
    
    for c in com:
        
        diff= [abs(i-j).astype('timedelta64[M]') for i in x[c[0]] for j in x[c[1]]]
        
        if 0< min (diff) <=3:
            
            valid.append(c)
    
    return valid
```
