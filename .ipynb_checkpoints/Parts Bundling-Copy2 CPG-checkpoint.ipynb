```python
import numpy as np
import pandas as pd
from collections import Counter, defaultdict
import time
from itertools import combinations, chain
```

```python
start= time.perf_counter()
```

```python
df= pd.read_csv('CPG Bundling data prep.csv', on_bad_lines='skip', sep= ',', encoding='latin-1')
```

```python
df.head()
```

```python
df.drop_duplicates().shape
```

```python
df.isna().sum()
```

```python
ecc_dict= dict(zip(df['Item Description'].astype(str)+ "|"+ df['Long Part Number'].astype(str), 
                   df[['ECC', 'BMS Product Code']].values.tolist()))
```

```python
len (ecc_dict)
```

```python
ecc_dict['FILTER-OIL (TEXT)|0122-0833']
```

```python
ecc_dict
```

```python
df['Invoice Number'].nunique()
```

```python
%%time

inv= df.groupby('Invoice Number')[['Item Description', 'Long Part Number']].agg(lambda x: "|".join(x)).reset_index()
```

```python
inv['Item Part']= inv[['Item Description', 'Long Part Number']].apply(lambda x: [i+"|"+j for i,j in 
                                                                            zip(x[0].split("|"), x[1].split("|"))], axis=1)
```

```python
inv_dict= dict(zip(inv['Invoice Number'], inv['Item Part']))
```

```python
inv
```

```python
part= df.groupby(['Item Description'])['Invoice Number'].apply(set).to_frame().reset_index()
```

```python
part
```

```python
part['len']= part['Invoice Number'].apply(len)
```

```python
#part['Item Part']= part['Item Description'].astype(str)+ "|" + part['Long Part Number'].astype(str)
```

```python
%%time

part['Counter']= part['Invoice Number'].apply(lambda x: Counter(chain(*(inv_dict[i] for i in x))).items())
```

```python
part['Len Final']= part['Counter'].apply(len)
```

```python
part_final= part[part['Len Final']>0]
```

```python
part_final.shape
```

```python
bun= part_final[['Item Description', 'len', 'Counter']].explode(['Counter']).reset_index(drop=True)
```

```python
bun['Associated Item']= bun['Counter'].apply(lambda x: x[0].split('|')[0])
```

```python
bun['Long Part Number']=  bun['Counter'].apply(lambda x: x[0].split('|')[1])
```

```python
bun['Part Occurence']= bun['Counter'].apply(lambda x: x[1])

bun['Association %age']= bun['Part Occurence']/bun['len']

#bun['Association %age']= np.round(bun['Association %age'], 2)
```

```python
bun['ECC']= bun['Counter'].apply(lambda x:ecc_dict.get(x[0], ["", ""])[0])
```

```python
bun['BMS Product Code']= bun['Counter'].apply(lambda x:ecc_dict.get(x[0], ["", ""])[1])
```

```python
bun
```

```python
bun['ECC'].isna().sum()
```

```python
bun['BMS Product Code'].isna().sum()
```

```python
bun_sub= bun[(bun['Item Description']!= bun['Associated Item']) & (bun['Association %age']>=0.10)]
```

```python
bun_sub.shape
```

```python
bun_sub
```

```python
bun_sub.drop(columns= ['Counter', 'len'], inplace= True)

#bun_sub.reset_index(inplace= True)
```

```python
## Add more columns to rename
```

```python
bun_sub.rename(columns= {
                        'Part Occurence': 'No. of Instances'}, inplace= True)
```

```python
bun_sub.to_csv('Bundling CPG.csv', index = False)
```

```python
end= time.perf_counter()

print('Time taken is', (end-start)/60, 'minutes')
```
